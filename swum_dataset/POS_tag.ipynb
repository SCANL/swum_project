```python
df = pd.read_csv("v_positional_freq.csv")
print(df)
```

```python
df1 = df[['word', 'first']]
df1.head(100)
```

```python
idName_list = df1['word'].head(100).tolist()
print (idName_list)
```

```python
import nltk
from nltk import word_tokenize
from nltk.corpus import wordnet   #Import wordnet from the NLTK
#syn = list()
#ant = list()

for word in idName_list:
    #print(word)
    syn = list()
    ant = list()
    for sysnset in wordnet.synsets(word):
        #print(sysnset.lemmas())
        for lem in sysnset.lemmas():
            #print(lem.name())
            syn.append(lem.name())    #add the synonyms
            if lem.antonyms():    #When antonyms are available, add them into the list
                ant.append(lem.antonyms()[0].name())
    text = word_tokenize(word)
    nltk.pos_tag(text)
    print(nltk.pos_tag(text))
    print(('Synonyms for {}: ' + str(syn)).format(word))
    print(('Antonyms for {}: ' + str(ant)).format(word))
```
